```python
# https://leetcode.com/problems/total-hamming-distance/description/

class Solution:
    def totalHammingDistance(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        ret = 0
        num_strs = zip(*[bin(n)[2:].rjust(30, '0') for n in nums]) # make groups of each bits
        
        for bit_group in num_strs:
            c1 = bit_group.count('1')
            ret += c1 * (len(bit_group) - c1) # the pairs of 1 and 0 makes the number of hamming distance
        return ret
```
